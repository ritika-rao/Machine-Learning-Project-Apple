###### Documentation
As Day percentage change increase volume of traded stocks also rise.
We conclude this by seeing that for almost all positive hills of Day percentage change, volume also has positive hills which reach heights more than others.Hence relation between Day percentage change and volume is directly propotional.

Inference: It implies that whenever the positive daily percentages increases our total traded quantity also rises by somewhat the same amount.
And since the daily changing percentage depends upon the closing prices of each day hence closing prices play a major in the increase of the total traded quantity

#### Query 9

```python
first=21
second=34

signals=pd.DataFrame(index=axis.index)
signals['signal']=0.0

#SMA of 21 Day
signals['small_mov_avg']=axis['Close Price'].rolling(window=first, min_periods=1,center=False).mean()

#SMA of 34 day
signals['long_mov_avg']=axis['Close Price'].rolling(window=second, min_periods=1,center=False).mean()
```

```python
signals['signal'][first:]=np.where(signals['small_mov_avg'][first:]>signals['long_mov_avg'][first:],1.0,0.0)

signals['positions']=signals['signal'].diff()
signals.head(50)
```

```python
fig = plt.figure(figsize=(14,9))
ax1 = fig.add_subplot(111, ylabel='Price')

#Plot the closing price
axis['Close Price'].plot(ax=ax1, color='black', lw=2.)

#plot the short and long moving averages
signals[['small_mov_avg', 'long_mov_avg']].plot(ax=ax1, lw=2.)

#Plot the buy signals
ax1.plot(signals.loc[signals.positions == 1.0].index, signals.small_mov_avg[signals.positions == 1.0], '>' , markersize=15,color='r')

#Plot the sell signals
ax1.plot(signals.loc[signals.positions == -1.0].index, signals.small_mov_avg[signals.positions == -1.0], 'v' , markersize=15)
plt.legend()
plt.show()
```

#### Query 10

```python
symbol = 'Axis Bank'

data = pd.read_csv('AXISBANK.csv'.format(symbol), index_col='Date',parse_dates=True, usecols=['Date', 'Close Price','Series'],na_values='nan')
data=data[data.Series == 'EQ']
data=data.drop(columns='Series')

# rename the column header with symbol name
data = data.rename(columns={'Close Price': symbol})
data.dropna(inplace=True)

# calculate Simple Moving Average with 14 days window
sma = data.rolling(window=14).mean()

# calculate the standar deviation
rstd = data.rolling(window=14).std()

upper_band = sma + 2 * rstd
upper_band = upper_band.rename(columns={symbol: 'upper'})
lower_band = sma - 2 * rstd
lower_band = lower_band.rename(columns={symbol: 'lower'})

data = data.join(upper_band).join(lower_band)


ax = data.plot(title='Axis Bank Price and Bollinger Bands',figsize=(14,9))
ax.fill_between(data.index, lower_band['lower'], upper_band['upper'], alpha='0.2',facecolor ='grey')
ax.set_xlabel('Date')
ax.set_ylabel('SMA and Bollinger Bands')
ax.grid()

plt.show()
```
